```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plotting graph
k=[1.1,2.3,4.5,10.11]
plt.plot(k)

#naming axis
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.show()
```

```python
#using dots

plt.plot([1,2,3,4],[1,3, 4,5],'ro')

#range of x and y axis
plt.axis([0, 6, 0, 6])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,3, 4,5])
plt.axis([0, 6, 0, 6])
plt.show()
```

```python

#generating random number
a=np.random.randn(6,4)
print(a)

#colorizing dots
plt.plot(a,a**2,'ro',a,a**3,'b^')
plt.axis([0,3,0,5])
plt.show()
```
