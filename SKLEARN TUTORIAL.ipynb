```python
import pandas as pd 
import numpy as np 
from sklearn import linear_model
from sklearn.model_selection import train_test_split

from sklearn.datasets import load_boston
boston = load_boston() #BOSTON HOUSING DATASET
```

```python
boston
```

```python
#CREATING DATAFRAME
df_x=pd.DataFrame(boston.data,columns=boston.feature_names)
df_y=pd.DataFrame(boston.target)
df_x.head() #see the data head
```

```python
#MAKING REGRESSION FUNCTION
reg = linear_model.LinearRegression()
```

```python
#SPLIT TRAIN AND TEST DATA, test to training data ratio = 0.2, then random the data
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=4)
```

```python
#now the data has been shuffled
x_train.head()
```

```python
#APPLYING REGRESSION TO TRAINING DATA
reg.fit(x_train,y_train)
```

```python
#WEIGHTS
reg.coef_
```

```python
# predict the house prices.
reg.predict(x_test)
```

```python
#mean square error 
np.mean((reg.predict(x_test) - y_test)**2)
```
